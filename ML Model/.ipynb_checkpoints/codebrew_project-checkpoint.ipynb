```python
import pandas as pd
import sqlite3
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# first things first, extract tables from database
conn = sqlite3.connect("../backend/database/reports.db")
```

```python
master_df = pd.read_sql_query("SELECT * FROM community_reports", conn)
master_df.head()
```

```python
master_df.dtypes
```

```python
# assuming we extract all 4 databases (MASTER, avg_df, known_val_feature_df, seasonal_constraints), preprocess master DB, THENNN (also preprocessing may be done backend, or maybe not)
# im doing preprocessing of master DB first
# so first removing un-needed columns:
print(master_df.columns.tolist())
master_df = master_df.drop(columns=["id", "community_name", "reporter_type", "seasonal_indicators_text", "notes"])
master_df.dtypes

# now we gotta convert everything to numeric
# these are report_date, local_season, rainfall_level, road_access, dog_roaming_level
# for local_season, gotta make maps tho:
season_map = {
    "Kudjewk": 1,
    "Bangkerreng": 2,
    "Yekke": 3,
    "Wurrkeng": 4,
    "Kurrung": 5,
    "Kunumeleng": 6
}

rainfall_map = {
    "high": 3,
    "medium": 2,
    "low": 1
}

road_access_map = {
    "difficult": 1,
    "moderate": 2,
    "easy": 3
}

dog_roaming_map = {
    "high": 3,
    "medium": 2,
    "low": 1
}

master_df["local_season"] = master_df["local_season"].map(season_map)
master_df["rainfall_level"] = master_df["rainfall_level"].map(rainfall_map)
master_df["road_access"] = master_df["road_access"].map(road_access_map)
master_df["dog_roaming_level"] = master_df["dog_roaming_level"].map(dog_roaming_map)

master_df.head()
```

```python
# now we gotta convert report_date into proper date format and make it id
master_df["report_date"] = pd.to_datetime(master_df["report_date"])
master_df = master_df.set_index("report_date")
master_df.head()
master_df.dtypes
```

```python
# LESGOOO Now there is one last preprocessing step. We need to make this into a month ting...
# So like make a month by month table for each of these tings.
monthly_df = master_df.drop(columns=["region_id"])
monthly_df = monthly_df.resample("MS").agg({
    "num_dogs_seen": "mean",
    "num_puppies_seen": "mean",
    "skin_issue_count": "mean",
    "recent_dog_deaths": "mean",
    "parasite_issue_count": "mean",
    "distance_to_clinic": "mean",
    "local_season": lambda x: x.mode().iloc[0],
    "rainfall_level": lambda x: x.mode().iloc[0],
    "road_access": lambda x: x.mode().iloc[0],
    "dog_roaming_level": lambda x: x.mode().iloc[0],
    "requested_help": lambda x: x.mode().iloc[0]
}).reset_index()

monthly_df.head()
```
